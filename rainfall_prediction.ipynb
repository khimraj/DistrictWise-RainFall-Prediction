```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

data = np.array(df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']])

X = None 
y = None

prev_months = 2

for i in range(data.shape[1] - prev_months):
    if X is None:
        X = data[:, i : i + prev_months]
        y = data[:, i + prev_months]
    else:
        X = np.concatenate((X, data[:, i : i + prev_months]), axis=0)
        y = np.concatenate((y, data[:, i + prev_months]), axis=0)
        
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rajasthan_data = df[['DISTRICT','JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL','AUG', 'SEP', 'OCT', 'NOV', 'DEC']].loc[df['STATE/UT'] == 'RAJASTHAN']
jalore_data = np.array(rajasthan_data[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL','AUG', 'SEP', 'OCT', 'NOV', 'DEC']].loc[rajasthan_data['DISTRICT'] == 'JALORE'])

jalore_X = None 
jalore_y = None

prev_months = 2

for i in range(jalore_data.shape[1] - prev_months):
    if jalore_X is None:
        jalore_X = jalore_data[:, i : i + prev_months]
        jalore_y = jalore_data[:, i + prev_months]
    else:
        jalore_X = np.concatenate((jalore_X, jalore_data[:, i : i + prev_months]), axis=0)
        jalore_y = np.concatenate((jalore_y, jalore_data[:, i + prev_months]), axis=0)
```

```python
#Function to plot the graphs
def plot_graphs(actual, predicted, title):
    fig = plt.figure(figsize=(13, 8))
    plt.plot(actual)
    plt.plot(predicted)
    plt.title(title, fontsize=15, weight='bold')
    plt.ylabel('RainFall in MM', fontsize=15, weight='bold')
    plt.xlabel('MONTH', fontsize=15, weight='bold')
    plt.xticks(range(10), ['MAR', 'APR', 'MAY', 'JUN', 'JUL','AUG', 'SEP', 'OCT', 'NOV', 'DEC'])
    plt.legend(['Actual','Predicted'], loc='upper right')
    plt.show()
    fig.savefig('Plot/' + title, dpi=300, bbox_inches='tight')
```

```python
from sklearn.linear_model import Lasso

# Lasso Regression model
reg = Lasso(alpha=0.1)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
mean_absolute_error(y_test, y_pred)
```

```python
jalore_y_pred = reg.predict(jalore_X)
mean_absolute_error(jalore_y, jalore_y_pred)
```

```python
plot_graphs(jalore_y, jalore_y_pred, 'Lasso Regression Model: Jalore RainFall Predication')
```

```python
import tensorflow as tf

# 1D CNN model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv1D(filters=32, kernel_size=1, activation='relu', input_shape=(prev_months,1)))
model.add(tf.keras.layers.Conv1D(filters=64, kernel_size=1, activation='relu'))
model.add(tf.keras.layers.Conv1D(filters=128, kernel_size=1, activation='relu'))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(1,activation = 'linear'))
opt = tf.keras.optimizers.Adam(lr=0.01, beta_1=0.9, beta_2=0.999)
model.compile(loss='mean_squared_error', optimizer=opt, metrics=['mae'])
model.summary()
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
model.fit(X_train, y_train, batch_size=32, epochs=10, validation_split=0.2, shuffle=True)
y_pred = model.predict(X_test)
mean_absolute_error(y_test, y_pred)
```

```python
jalore_X = jalore_X.reshape(jalore_X.shape[0], jalore_X.shape[1], 1)
jalore_y_pred = model.predict(jalore_X)
mean_absolute_error(jalore_y, jalore_y_pred)
```

```python
plot_graphs(jalore_y, jalore_y_pred, '1D CNN Model: Jalore RainFall Predication')
```

```python
# LSTM model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.LSTM(64, activation='relu', return_sequences=True, input_shape=(prev_months,1)))
model.add(tf.keras.layers.LSTM(32, activation='relu'))
model.add(tf.keras.layers.Dense(1))
opt = tf.keras.optimizers.Adam(lr=0.01, beta_1=0.9, beta_2=0.999)
model.compile(loss='mean_squared_error', optimizer=opt, metrics=['mae'])
model.summary()
```

```python
model.fit(X_train, y_train, batch_size=32, epochs=10, validation_split=0.2, shuffle=True)
y_pred = model.predict(X_test)
mean_absolute_error(y_test, y_pred)
```

```python
jalore_y_pred = model.predict(jalore_X)
mean_absolute_error(jalore_y, jalore_y_pred)
```

```python
plot_graphs(jalore_y, jalore_y_pred, 'LSTM Model: Jalore RainFall Predication')
```
